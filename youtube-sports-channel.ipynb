```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
sports = pd.read_csv("/kaggle/input/youtube-sports-channels-statistics/yt_sports_channels_stats.csv")
```

```python
# Rename the column 
sports.rename(columns={'channel_title': 'channel name'}, inplace=True)
```

# Analysis by video count

```python
top_20_channel_by_video = sports.groupby('channel name')['video_count'].sum().reset_index().sort_values(by='video_count', ascending = False)

top_20_channel_by_video.head(20)
```

```python
# Create a dynamic Pie chart
top_50=top_20_channel_by_video.head(50)
fig = px.pie(top_50, values='video_count', names='channel name', 
             title='channels Distribution by video count',
             hover_data=['video_count'], 
             labels={'video count': 'video_count'}, 
             hole=0.6,
             width=800, 
             height=600)

# Show the plot
fig.show()
```

```python
fig = px.bar(top_50, x='channel name', y='video_count', 
             title='channels Distribution by video count',
             color='channel name',
             hover_data=['video_count'], 
             labels={'video_count': 'video_count'}, 
             width=1100,  
             height=700)  

# Show the plot
fig.show()
```

```python
# Convert view count to crore
sports['view_count'] = sports['view_count'] / 10000000
```

```python
# Rename the column 
sports.rename(columns={'view_count': 'view count (crores)'}, inplace=True)
```

# Analysis by view count

```python
top_20_channel_by_view = sports.groupby('channel name')['view count (crores)'].sum().reset_index().sort_values(by='view count (crores)', ascending = False)

top_20_channel_by_view.head(20)
```

```python
# Create a dynamic Pie chart
top_50=top_20_channel_by_view.head(50)
fig = px.pie(top_50, values='view count (crores)', names='channel name', 
             title='channels Distribution by view count (crores)',
             hover_data=['view count (crores)'], 
             labels={'view count (crores)': 'view count (crores)'}, 
             hole=0.6,
             width=800, 
             height=600)

# Show the plot
fig.show()
```

```python
fig = px.bar(top_50, x='channel name', y='view count (crores)', 
             title='channels Distribution by view count (crores)',
             color='channel name',
             hover_data=['view count (crores)'], 
             labels={'view count (crores)': 'view count (crores)'}, 
             width=1100,  
             height=700)  

# Show the plot
fig.show()
```

```python
# Convert view count to crore
sports['subscriber_count'] = sports['subscriber_count'] / 10000000
```

```python
# Rename the column 
sports.rename(columns={'subscriber_count': 'subscriber count (crore)'}, inplace=True)
```

# Analysis by subscriber count

```python
top_20_channel_by_subscriber = sports.groupby('channel name')['subscriber count (crore)'].sum().reset_index().sort_values(by='subscriber count (crore)', ascending = False)

top_20_channel_by_subscriber.head(20)
```

```python
# Create a dynamic Pie chart
top_20=top_20_channel_by_subscriber.head(20)
fig = px.pie(top_20, values='subscriber count (crore)', names='channel name', 
             title='channels Distribution by subscriber count (crore)',
             hover_data=['subscriber count (crore)'], 
             labels={'subscriber count (crore)': 'subscriber count (crore)'}, 
             hole=0.6,
             width=800, 
             height=600)

# Show the plot
fig.show()
```

```python
fig = px.bar(top_20, x='channel name', y='subscriber count (crore)', 
             title='channels Distribution by subscriber count (crore)',
             color='channel name',
             hover_data=['subscriber count (crore)'], 
             labels={'subscriber count (crore)': 'subscriber count (crore)'}, 
             width=1100,  
             height=700)  

# Show the plot
fig.show()
```
